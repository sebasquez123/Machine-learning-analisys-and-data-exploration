## Conceptualización base

*MODELOS DE CLASIFICACIÓN*
- Regresión Logística: modelo de aprendizaje supervisado utilizado para resolver problemas de clasificación, especialmente binaria. utiliza la función Sigmoide para a convertir las predicciones. su salida es binaria de **clase 1 = 1 si la probabilidad >= 0.5** o **clase 0 = 0 si la probabilidad < 0.5** 

*MODELO BASADO EN ARBOLES DE DESICIÓN*
- se utiliza tanto para problemas de clasificación como de regresión. Es fácil de interpretar y visualiza cómo el modelo toma decisiones. Un modelo basado en toma de decisiones por medio de reglas que define su propagabilidad por la probabilidad y la entropía de los datos. Salida de clasificación o regresión devolviendo la clase más frecuente o el promedio de las hojas.

*NEIVE BAYES*
- basado en el Teorema de Bayes y es utilizado principalmente para problemas de clasificación. Su característica principal es la suposición de independencia condicional entre las características. Calcula las probabilidades de que una muestra pertenezca a cada clase (0 o 1). Luego, asigna la clase con la probabilidad más alta como la predicción final. si su salida es 0.75, asigna un 1.
    - Gaussiano: Para datos continuos, asume que las características siguen una distribución normal.
    - Multinomial: Ideal para datos discretos, como conteos (e.g., palabras en texto).
    - Bernoulli: Trabaja con variables binarias (e.g., presencia o ausencia de una palabra)

*MAQUINAS DE SOPORTE VECTORIAL*
- Su objetivo es encontrar un hiperplano que separe las clases de datos de la manera más clara posible, maximizando el margen entre las clases.
Una vez entrenado, el modelo predice la clase de los nuevos puntos de datos basándose en qué lado del hiperplano se encuentran. Es muy efectivo en clasificación binaria y puede extenderse a problemas multi clase mediante técnicas como One-vs-One o One-vs-All.
El modelo devuelve las clases predichas para cada observación, 0 o 1



## Evaluación y desempeño
- Sobre ajuste y sub ajuste:

Se evalúan los 4 conjuntos de datos que se segmentaron en el inicio, con el fin de identificar la exactitud de la respuesta en el orden de 0 a 1, y validar el desempeño del modelo frente a los datos nuevos. se entrena con los datos de entrenamiento y se compara con los mismos para identificar si responde bien a los datos estudiados, así mismo se vuelve y se evalúa con los datos de prueba y se compara con los mismos. Si la exactitud o rendimiento con los datos ya estudiados es mayor a los datos nuevos, se considera que el modelo esta sobre ajustado, es decir no generaliza eventos nuevos. si el rendimiento es bajo, tanto para datos ya estudiados como para los nuevos, el modelo tiene un subajuste que habrá que tratar con:

    - Ajuste de hiperparametros
    - Aumento de cantidad y calidad de los datos de entrenamiento
    - Aumento del tiempo de entrenamiento

-  Validación Cruzada:

una técnica utilizada para evaluar el rendimiento de un modelo de aprendizaje automático y garantizar que no haya sobreajuste (overfitting) o subajuste (underfitting) al conjunto de datos de entrenamiento. La validación cruzada divide el conjunto de datos en varias partes y entrena el modelo varias veces, utilizando una de estas partes como conjunto de prueba (test) y las demás como conjunto de entrenamiento (train). Esto se repite para cada una de las partes. Al final, el rendimiento del modelo se promedia a lo largo de todas las iteraciones, lo que proporciona una estimación más robusta y generalizada del desempeño. el resultado de validación puede ser similar a la técnica de sub ajuste y sobre ajuste manual.

- Reporte de clasificación:

proporciona un informe detallado sobre el rendimiento de un modelo de clasificación:

**Precisión (Precisión):**
Indica la exactitud de las predicciones positivas (CLASE 1). Es decir, de todas las veces que el modelo predijo una clase positiva, cuántas fueron correctas.
Interpretación: Alta precisión significa que las predicciones positivas son correctas en su mayoría.

**Recall (Sensibilidad o Exhaustividad):**
Mide cuántas veces el modelo identificó correctamente las muestras positivas. Es decir, de todas las instancias que eran realmente positivas, cuántas fueron predichas correctamente. 
Interpretación: Un alto recall significa que el modelo es muy bueno detectando la clase positiva, aunque pueda haber cometido más falsos positivos.

**F1-Score:**
Es la media armónica de la precisión y el recall. Se usa cuando se necesita un equilibrio entre ambas métricas.
Interpretación: Un buen F1-score indica que el modelo tiene un buen balance entre precisión y recall.

**Support:**
El número de instancias reales de cada clase en el conjunto de prueba.
Interpretación: Muestra cuántas observaciones pertenecen a cada clase.

- Matriz de confusión:

Esta matriz proporciona una representación visual y cuantitativa de los resultados de clasificación, comparando las predicciones del modelo con los valores reales (o verdaderos).  

a. True Positives (TP): Casos que el modelo predijo correctamente como positivos. Son los casos en los que el modelo acertó al predecir la clase positiva.

b. True Negatives (TN): Casos que el modelo predijo correctamente como negativos. Son los casos en los que el modelo acertó al predecir la clase negativa.

c. False Positives (FP): Casos que el modelo predijo incorrectamente como positivos, pero en realidad eran negativos. También conocidos como errores tipo I.

d. False Negatives (FN): Casos que el modelo predijo incorrectamente como negativos, pero en realidad eran positivos. También conocidos como errores tipo II.


## Análisis y visualización de tendencia de rendimiento



- **Curva ROC: **

La Curva ROC es una representación gráfica que ilustra el rendimiento de un clasificador binario.
En el eje Y, se encuentra el True Positive Rate (TPR), también conocido como Recall, que mide cuántos casos positivos fueron identificados correctamente. En el eje X, se encuentra el False Positive Rate (FPR), que mide cuántos casos negativos fueron incorrectamente clasificados como positivos
Una gráfica ideal sería la que termina en la parte superior izquierda, donde el recall es alto y al FPR es bajo.

- **AUC (Area Under the Curve):**

AUC = 1: Modelo perfecto (sin errores).  
AUC = 0.5: Modelo aleatorio, sin capacidad de clasificación.  
AUC < 0.5: Modelo peor que aleatorio (lo que indicaría que el modelo tiene que ser corregido).

## Identificación de datos irrelevantes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("electric_cars.csv")
data = pd.DataFrame(data)
relevante = ['Battery Capacity (kWh)','Energy Consumed (kWh)','Charging Duration (hours)','Charging Rate (kW)','Time of Day','Day of Week','State of Charge (Start %)','State of Charge (End %)','Distance Driven (since last charge) (km)','Temperature (°C)','Vehicle Age (years)','Vehicle Model','User Type']

data= data[relevante]
data = data[data['User Type'].isin(['Commuter','Long-Distance Traveler'])]
data.head()
```

## Detección y manejo de datos faltantes

```python

data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
print(data.isnull().sum())
print(data.info())
```

## Categorizar caracteristicas

```python
data['User Type'].replace({'Commuter':1,'Long-Distance Traveler':0},inplace=True)
data['Time of Day'].replace({'Morning':1,'Evening':2,'Afternoon':3,'Night':4},inplace=True)
data['Day of Week'].replace({'Tuesday':2,'Thursday':4, 'Saturday':6, 'Friday':5, 'Monday':1, 'Wednesday':3,'Sunday':7},inplace=True)    
data['Vehicle Model'].replace({'BMW i3':1, 'Chevy Bolt':2, 'Hyundai Kona':3, 'Nissan Leaf':4,'Tesla Model 3':5},inplace=True)
data
```

```python
data = (data-data.min())/(data.max()-data.min())
data.head()
```

## Divición en conjuntos de datos (HOLD OUT)

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.metrics import roc_curve, auc, RocCurveDisplay
from sklearn.linear_model import LogisticRegression

ind= ['Battery Capacity (kWh)','Energy Consumed (kWh)','Charging Duration (hours)','Charging Rate (kW)','Time of Day','Day of Week','State of Charge (Start %)','State of Charge (End %)','Distance Driven (since last charge) (km)','Temperature (°C)','Vehicle Age (years)','Vehicle Model']
x= data[ind]
y=data['User Type']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```
