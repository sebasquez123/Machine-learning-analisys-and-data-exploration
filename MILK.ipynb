**Definición del objetivo y analisis de viabilidad**

La identificación del problema en Machine Learning es el primer paso. Implica definir claramente el desafío que se requiere solventar, entendiendo de primera mano su importancia y cómo el aprendizaje automático puede ser útil.

- Especificar el desafío de forma medible, es decir que se pretende predecir.

- Asegurarse de que el problema es aplicable para técnicas de Machine Learning, es decir, si hay patrones en los datos que se pueden aprender para automatizar la toma de decisiones.
- Determinar si se trata de un problema de clasificación, regresión, clustering, etc.

**Evaluación, métricas e indices de desempeño**

Finalmente, en la evaluación del modelo, se miden métricas como precisión, recall y F1-score para determinar si el modelo es lo preciso, apto y acorde a lo esperado para ser desplegado.

#  APLICACION DE MODELO
**Para predecir la participación diaria de cada municipio en el departamento del valle del cauca**, se requiere emplear 3 modelos (regresión lineal multiple-Arbol de decisión-Maquinas de soporte Vectorial) de regresión lineal que afiancen el comportamiento indicado para manejar un rango de predicción aceptable.


```python
data= datatotrain
data.head()
data.reset_index(drop=True, inplace=True)
data.head()
```

## Divición en conjuntos de datos (HOLD OUT)

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
col= ['Número de  Vacas en Ordeño','Producción  diaria (Litros)','Producción Promedio Vaca dia (Litros)']
x= data[col]
y= data['Participación en la Producción  Diaria de Leche  (%)']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

**Una vez divididos los datos de entrenamiento y prueba, para efectuar un buen rendimiento de entrenamiento del modelo, se realiza la normalización o estandarización de los conjuntos**, en la sección de análisis de datos se produjo una normalización para la visualización, sin embargo no es correcto, debido a que para efectuar el método, se debe contar con información de todo el dataset, lo cual afectaría el conjunto de datos de entrenamiento que tendrían en si, una relación con el conjunto de datos de prueba. para evitar fuga de datos en el entrenamiento, primero se dividen y luego se normalizan.

*Nota:* La salida Y_test y Y_train deben mantenerse en su formato original para que el modelo las pueda aprender correctamente. La normalización de las etiquetas puede alterar el objetivo y afectar negativamente el rendimiento del modelo.


## _REGRESIÓN LINEAL MULTIPLE_

```python

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

model = LinearRegression()
model.fit(x_train, y_train)

prediccion = model.predict(x_test)


### VALIDACIÓN DE SUB Y SOBRE AJUSTE  ###

y_train_pred = model.predict(x_train)
train_accuracy = mean_absolute_error(y_train, y_train_pred)
print("Train MSE:", train_accuracy)

y_test_pred = model.predict(x_test)
test_accuracy = mean_absolute_error(y_test, y_test_pred)
print("Test MSE:", test_accuracy)

### VALIDACIÓN CRUZADA ###

cv_scores = cross_val_score(model, x, y, cv=5, scoring='r2')
print("CV R2: ", cv_scores.mean())

### EVALUACIÓN ###

r2 = r2_score(y_test, prediccion)
print(f"R²: {r2}")
```

| MAE Train | MAE Test | R2 CV | R2 | 
|-----------|-----------|-----------|-----------|
| 0.0049    | 0.0071    | 0.3078    | 0.9999    |

## _ARBOL DE DECISIÓN_

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import cross_val_score


model = DecisionTreeRegressor(criterion='poisson', max_depth= 5, max_features= None, min_samples_leaf= 1, min_samples_split= 2)

model.fit(x_train, y_train)
prediccion = model.predict(x_test)


### VALIDACION DE SUB Y SOBRE AJUSTE  ###


y_train_pred = model.predict(x_train)
train_accuracy = mean_absolute_error(y_train, y_train_pred)
print("Train MSE:", train_accuracy)

y_test_pred = model.predict(x_test)
test_accuracy = mean_absolute_error(y_test, y_test_pred)
print("Test MSE:", test_accuracy)

### VALIDACIÓN CRUZADA ###

cv_scores = cross_val_score(model, x, y, cv=5, scoring='r2')
print("CV R2: ", cv_scores.mean())

### EVALUACIÓN ###
r2 = r2_score(y_test, prediccion)
print(f"R²: {r2}")

### GRAFICO DE PROFUNDIDAD ###

plt.figure(figsize=(20, 10))
plot_tree(model, feature_names=x.columns, filled=True)
plt.show()
```

| MAE Train | MAE Test | R2 CV | R2 | 
|-----------|-----------|-----------|-----------|
| 0.0125    | 0.4216    | 0.9260    | 0.8724    |

## _MAQUINAS DE SOPORTE VECTORIAL_
Usando la libreria de GridSearchCV o RandomizedSearchCV podemos buscar los mejores hiperparametros despues de un analisis

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1.0, 10, 100],
    'kernel': ['rbf', 'linear', 'poly'],
    'gamma': ['scale', 'auto', 0.01, 0.1],
    'epsilon': [0.01, 0.1, 0.2, 0.5],
}

# Crear el modelo SVR
svr = SVR()

# Búsqueda de grid con validación cruzada
grid_search = GridSearchCV(svr, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(x_train, y_train)

# Mejor configuración de hiperparámetros
print("Mejores hiperparámetros:", grid_search.best_params_)

# Usar el mejor modelo
best_model = grid_search.best_estimator_

# Evaluación con el mejor modelo
prediccion = best_model.predict(x_test)

### VALIDACION DE SUB Y SOBRE AJUSTE  ###


y_train_pred = best_model.predict(x_train)
train_accuracy = mean_absolute_error(y_train, y_train_pred)
print("Train MAE:", train_accuracy)

y_test_pred = best_model.predict(x_test)
test_accuracy = mean_absolute_error(y_test, y_test_pred)
print("Test MAE:", test_accuracy)

### VALIDACIÓN CRUZADA ###
cv_scores = cross_val_score(model, x, y, cv=5, scoring='neg_mean_absolute_error')
print("CV MAE: ", -cv_scores.mean())  # Convertir el valor negativo a positivo

### EVALUACIÓN ###
r2 = r2_score(y_test, prediccion)
print(f"R²: {r2}")
```
